# MCQ Quiz

```python
#Questions of Quiz
Q = {"Number of Continents on Earth":"1:Seven 2:Four 3:Five 4:Six",
    "Which country successfully landed a man on moon":"1:USA 2:Germany 3:France 4:England",
    "Which country is not present in the continent of North America":"1:Mexico 2:USA 3:Cannada 4:Brazil",
    "Which is the American National Bird":"1:Parrot 2:Crane 3:Eagle 4:Swan",
    "Which country shares the longest border with USA": "1: Mexico 2:Brazil 3:Cannada 4:Argentina",
    "Which is the financial capital of USA":"1:Sanfrancisco 2:Boston 3:Los Angeles 4:NYC",
    "Which of the following space agencies belongs to the USA":"1:BNSC 2:ISRO 3:DLR 4:NASA",
    "Which is the second most spoken language in USA": "1:English 2:Spanish 3:Mandarin 4:Italian",
    "Which is the capital city of USA":"1:NYC 2:Washington DC 3:Austin 4:San Jose",
    "Mt Rushmore has the sclupture of":"1:Former US Presidents 2:Hollywood stars 3:Military Personel 4:Scientists"}

#Options of all the questions
Opt = list(Q.values())

#All the corect answeres
ans = [1,1,4,3,3,4,4,2,2,1]
```

```python
print("The MCQ Quiz will have 10 questions with 4 options in each.")
print("Right answers will give you 2 marks and wrong answer will take away 1 mark")
print("If selection is made out of choices, no marks will be awarded")
print("\n"*2)

score = 0 #will keep track of user's score in quiz
c = 0
p = iter(Opt)
for i in Q:
    print(i)
    a = next(p)
    print(a)
    print("\n")
    selection = int(input("Enter the option between 1 and 4: ")) #taking user selection from options
    print("\n")
    
    if selection in (1,2,3,4):
        
        if selection == ans[c]:
            score = score + 2  #user gives right answer hence +2 marks are awarded
        else:
            score = score - 1  #user gives wrong answer hence 1 mark is deducted
    
    c = c + 1

print("The player has scored",score,"marks")
```

```python
#Improvements done in previous code:
#Added 5th option to skip the question
# Negative marking of 1 marks applicable for wrong choices
#Entering user in an infinite while loop until he enters a valid option (from 1 to 5)
#Made the code exception safe by first checking if the option is valid and then converting to int.
#Other logics are same as previous code

import random
question_bank = [
        ["Which one of the following is not a keyword in Python language?","pass","eval","assert","nonlocal",2],
        ["How is a code block indicated in Python?","Parantheses","Square Brackets","Indentation","Space",3],
        ["Which symbol is used for the modulus operator in Python?","%","&","**","^",1],
        ["What is the result of 2 + 3 * 5 in Python?","25","15","17","13",3],
        ["Which data type is used to store a sequence of characters in Python?","char","int","float","str",4],
        ["What is the output of len(\"Python\") ?","5","6","7","Error",1],
        ["Which of the following is not a valid Python data type in python?","bool","complex","decimal","character",4],
        ["What is the purpose of the if statement in Python?","To define a function","To loop through a sequence of items","To perform conditional execution","To calculate mathematical expressions",3],
        ["Which of the following data types in Python is mutable?","int","list","float","tuple",2  ],
        ["In Python, which data structure uses a key-value pair for storage?","List","Set","Dictionary","Tuple",3]
    ]
random.shuffle(question_bank)
marks=0
for question in question_bank:
    print(question[0])
    for i in range(1,5):
        print(i,":",question[i])
    print("5 : Skip Question")
    choice = input("Your Answer: ")
    while(choice not in ["1","2","3","4","5"]):
        print("Error: Incorrect input!")
        choice = input("Your Answer: ")
    choice = int(choice)
    if choice == question[5]:
        marks+=2
    elif choice != 5:
        marks-=1
print("You scored",marks,"out of 20 marks")
```
