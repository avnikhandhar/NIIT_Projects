```python
#dictionary to keep record of quantity of available items
avail_quantity= {"Biscuits":5,"Cereals":10,"Chicken":20,"Oats":20,"Rice":30}

#dictionary to keep record of prices of available items
cost = {"Biscuits":2.5,"Cereals":3.5,"Chicken":13.3,"Oats":3.9,"Rice":4.5}

#list to save the order details
order = []

#user defined function to generate bill
def bill_generation(n,ad,deli_rate):
    
    #giving the headings of the bill
    top = "   BILL"
    y = top.center(50)
    print(y)
    print("\n")
    print("SNo"+"\t\t"+"   Item"+"\t\t  "+"Quantity"+"\t"+"Total Cost")
   
    l2 = len(order)
    
    #printing the bill
    for i in range(0,l2):
        print(i,"\t\t",order[i][0],"\t\t",order[i][1],"\t\t",order[i][2])
        
    #initialising the bill amount to zero
    bill = 0
    
    #adding price of every item as per the purchase
    for i in range(0,l2):
        bill = bill + order[i][2]
    
    print("\n") #to make the output more readable
    
    #printing the total bill amount
    print("Total Item Cost=Total Item Cost + Delivery Charges=$",bill,"+",deli_rate,"=$",bill+deli_rate)
    
    #printing customer details
    print(n)
    print(ad)

#user defined function to calculate the delivery charges
def delivery_charges(d):
    if d>=0 and d<5:
        dc = 10
    elif d>=5 and d<10:
        dc = 12
    elif d>=10 and d<15:
        dc = 14
    else:
        dc = 20
    
    return dc

#user defined function to store the customer details for delivery
def customer_details():
    print("Fill in your delivery details:")
    
    name = input("Enter your name ")
    address = input("Enter your address ")
    distance = int(input("Enter the distance from store(Choose from options):\n1.0-5 miles\n2.5-10 miles\n3.10-15 miles\n4.Above 15 miles "))
    
    deli_charges = delivery_charges(distance)
    
    bill_generation(name,address,deli_charges)


#user defined function to take order from user
def take_order():
    
    #list to save the order details of each item
    s = []
    
    #taking order for each item individually
    bis = int(input("How many packets of biscuits are required?"))
    cer = int(input("How many packets of cereals are required?"))
    chi = int(input("How many packets of frozen chicken are required?"))
    oat = int(input("How many packets of oats are required?"))
    ric = int(input("How many packets of rice are required?"))
    
    #checking for 'Biscuits' item
    if bis>avail_quantity["Biscuits"]:
        print("Sorry! Do not have enough stock for biscuits")
    else:
        avail_quantity["Biscuits"] = avail_quantity["Biscuits"] - bis
        p = cost["Biscuits"] * bis
        s.append("Biscuits")
        s.append(bis)
        s.append(p)
        order.append(s)
        s = []
    
    #checking for 'Cereals' item
    if cer>avail_quantity["Cereals"]:
        print("Sorry! Do not have enough stock for cereals")
    else:
        avail_quantity["Cereals"] = avail_quantity["Cereals"] - cer
        p = cost["Cereals"] * cer
        s.append("Cereals")
        s.append(cer)
        s.append(p)
        order.append(s)
        s = []
    
    #checking for 'Frozen Chicken' item
    if cer>avail_quantity["Chicken"]:
        print("Sorry! Do not have enough stock for frozen chicken")
    else:
        avail_quantity["Chicken"] = avail_quantity["Chicken"] - chi
        p = cost["Chicken"] * chi
        s.append("Chicken")
        s.append(chi)
        s.append(p)
        order.append(s)
        s = []
        
    #checking for 'Oats' item
    if oat>avail_quantity["Oats"]:
        print("Sorry! Do not have enough stock for oats")
    else:
        avail_quantity["Oats"] = avail_quantity["Oats"] - oat
        p = cost["Oats"] * oat
        s.append("Oats")
        s.append(oat)
        s.append(p)
        order.append(s)
        s = []
        
    #checking for 'Rice' item
    if ric>avail_quantity["Rice"]:
        print("Sorry! Do not have enough stock for rice")
    else:
        avail_quantity["Rice"] = avail_quantity["Rice"] - ric
        p = cost["Rice"] * ric
        s.append("Rice")
        s.append(ric)
        s.append(p)
        order.append(s)
        s = []
        
    #giving confirmation on order
    print("Order taken successfully!")
    
    #calling a user defined function to get the customer details for delivery
    customer_details()

#user defined function to show the available items with their available quantity and prices
def disp_cart():
    head = "ESSENTIAL ITEMS AVAILABLE"
    x = head.center(50)
    print(x)
    print("\n")  
    print("SNo"+"\t\t"+"Item"+"\t\t\t"+"Quantity"+"\t"+"Cost/Item")

    l = [[1,"Biscuits: a pack of 20",5,2.50],
        [2,"Cereals: 2 pounds/pack",10,3.50],
        [3,"Chicken: 1 pound/pack",20,13.30],
        [4,"Oats: 4 pounds/pack",20,3.90],
        [5,"Rice: 2 pounds/pack",30,4.50]]

    for i in l:
        print(i[0],"\t"+i[1],"\t\t  ",i[2],"\t\t  ",i[3])
        
    take_order()

def main():
    disp_cart()
    
if __name__=="__main__":
    main()
```
